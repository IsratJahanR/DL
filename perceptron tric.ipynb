```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
print(X)
```

```python
print(y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def step(x):
    if x>0: return 1
    else: return 0
```

```python
def perceptron(X,y):
    
    X = np.insert(X,0,1,axis=1) # insert a column with all value 1 into X-0th index
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(1000):# hare epoc = 1000
        # J : indicate that withing 100 students , which one is taken
        j = np.random.randint(0,100) 
        y_hat = step(np.dot(X[j],weights))  # current output
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
    return weights[0],weights[1:]
      
```

```python
intercept, coef = perceptron(X,y)
```

```python
print(coef)
print(intercept)
```

```python
m = -(coef[0]/coef[1])
c = -(intercept/coef[1])
```

```python
print(m," ",c)
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c
```

```python
print(x_input)
print(y_input)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
